Combining a larger batch size (64) with a fast epsilon decay (5%) to see

```python
!python src/train.py --run_name expB_10 --timesteps 500000 --batch_size 64 --eps_start 1.0 --eps_end 0.02 --eps_decay_frac 0.05
```

## Conclusion

Across all 10 experiments, **batch size had the single largest impact** on Pong performance. 
Increasing the batch size from 32 (baseline) to 128 improved final mean reward from -20.8 to 
**-12.1 (expB_4)** — by far the strongest result in the set. Batch=64 (expB_3, expB_10) also 
outperformed the baseline substantially, reaching -15.0 and -15.6 respectively. In contrast, 
reducing batch size to 16 (expB_2) gave no improvement over baseline, suggesting smaller 
batches produce noisier gradient estimates that slow learning in this environment.

Epsilon schedule changes had a more moderate effect. Lowering the exploration floor to 0.01 
(expB_5) and using a fast decay of 2% (expB_7) or slow decay of 30% (expB_8) each gave modest 
improvements over baseline (roughly -15.7 to -17.2), showing that Pong is fairly tolerant of 
different exploration schedules as long as the agent eventually commits to exploitation.

The two negative-control configurations behaved exactly as predicted: raising the exploration 
floor to 0.2 (expB_6) and starting epsilon lower at 0.5 (expB_9) both stayed flat near -20.8, 
confirming that excessive or premature exploration prevents the agent from exploiting what 
it has learned.

**Best overall configuration: expB_4** (batch_size=128, eps_start=1.0, eps_end=0.05, 
eps_decay_frac=0.1), reaching a mean reward of -12.1 after 500,000 timesteps — the closest 
any run came to competitive Pong play. This suggests that for DQN on Pong, prioritizing a 
larger batch size gives more reliable gains than fine-tuning the exploration schedule alone, 
likely because larger batches reduce variance in the Q-value updates and lead to more stable 
convergence.
